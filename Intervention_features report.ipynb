# Report

### How early participants look at each intervention (timetofirstfixation)

- timetofirstfixation: how long until user looks at the intervention (from the start of the task)
- trigger_delay: how long until intervention is triggered (from the start of the task)
- how early participants look at each intervention: timetofirstfixation - trigger_delay

```python
import pandas as pd
import numpy as np
from IPython.display import display
pd.options.display.max_rows = None
import matplotlib.pyplot as plt
```

```python
# loading data

# all the relevant data integrated in one place
all_segs = pd.read_csv("Results\integrated_data.csv", index_col=0)

# all_segs_oredered_time = pd.read_csv("Results\\all_segs_oredered_time.csv", index_col=0)

# trigger_to_fixation_times
trigger_to_fixation_times = pd.read_csv("Results\\trigger_to_fixation_times.csv", index_col=0)

# trigger_to_fixation_times_participant
trigger_to_fixation_times_participant = pd.read_csv("Results\\trigger_to_fixation_times_participant.csv", index_col=0)

# proportion_counts
proportion_counts = pd.read_csv("Results\\proportion_counts.csv", index_col=0)

# proportion_counts
proportion_stats = pd.read_csv("Results\\proportion_stats.csv", index_col=0)
```

# Mean and STD of proportions 
- (over all msnv documents for each participant)

```python
display(proportion_stats.head(10))
```

```python
# proportion_stats['mean processed/triggered'].hist();
# plt.title('processed/triggered ratio'); plt.ylabel('frequency'); plt.xlabel('mean [processed/triggered ratio]');
bar_df = proportion_stats
pt = bar_df[['mean processed/triggered', 'std processed/triggered']].sort_values(by='mean processed/triggered')
pr = bar_df[['mean processed/total', 'std processed/total']].sort_values(by='mean processed/total')
tt = bar_df[['mean triggered/total', 'std triggered/total']].sort_values(by='mean triggered/total')

pt.plot.bar(figsize=(15,5));
pr.plot.bar(figsize=(15,5));
tt.plot.bar(figsize=(15,5));
```

```python
print("processed/triggered")
print("\t processed/triggered grand mean: ", proportion_stats['mean processed/triggered'].mean())
print("\t processed/triggered grand sdt:"  , proportion_stats['mean processed/triggered'].std())

print("processed/total")
print("\t processed/total grand mean: ", proportion_stats['mean processed/total'].mean())
print("\t processed/total grand sdt:"  , proportion_stats['mean processed/total'].std())

print("triggered/total")
print("\t triggered/total grand mean: ", proportion_stats['mean triggered/total'].mean())
print("\t triggered/total grand sdt:"  , proportion_stats['mean triggered/total'].std())
```

```python
# display(trigger_to_fixation_times_participant)

print("trigger_to_fixation_times")
print("\t trigger_to_fixation_times grand mean: ", trigger_to_fixation_times_participant['mean trigger_to_fixation_time'].mean())
print("\t trigger_to_fixation_times grand sdt:"  , trigger_to_fixation_times_participant['mean trigger_to_fixation_time'].std())
```
